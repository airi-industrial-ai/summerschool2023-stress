```python
data_path = '../data/'
```

# ECL

```python
ecl = pd.read_csv(data_path + 'ECL.csv', index_col=0)
ecl.index = pd.to_datetime(ecl.index) 
```

```python
print(ecl.shape)
ecl.head()
```

#### 321 features, so let's check last 10:

```python
for col in ecl.columns[-10:]:
    plt.figure(figsize=(20, 2))
    ecl[col].iloc[:500].plot()
    plt.title(col)
    plt.xlabel('')
    plt.show()
```

```python
# plt.figure(figsize=(20, 10))
for col in ecl.columns.values[-10:]:
    autocorrelation_plot(ecl[col][500:700], label=col)
plt.legend(loc=(1.04, 0));
```

```python
corr_ecl = ecl[ecl.columns.values[-10:]].corr()
corr_ecl.style.background_gradient(cmap='coolwarm')
```

```python
ecl['mov_avg'] = ecl.MT_320.rolling(24).mean()

plt.figure(figsize=(20, 2))
ecl['MT_320'].iloc[:1000].plot()
ecl['mov_avg'].iloc[:1000].plot(linewidth=5)
plt.title('MT_320')

plt.xlabel('')
plt.show()
```
